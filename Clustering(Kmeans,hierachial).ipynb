# Machine Learning Process:

```python
# identifying X
X = data_copy.iloc[:,:]
X.head()
```

```python
X.shape
```

```python
# performing scaling because there is a difference in the magnitudes of features:
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

numerical_features = [col for col in data_copy.columns if data_copy[col].dtype=='int']
categorical_features = [ col for col in data_copy.columns if data_copy[col].dtype!='int']

scaled_numerical_features = scaler.fit_transform(data_copy[numerical_features])
#converting scaled features from array to dataframe
numerical_features_df = pd.DataFrame(scaled_numerical_features,columns=numerical_features)

#merging the scaled columns with encoded columns for further machine learning process
final_data = pd.concat([numerical_features_df,data_copy[categorical_features]],axis=1)
```

```python
final_data.head()
```

- We have not performed scaling on columns which have 'unit8' datatype because no need to perform scaling as we have converted the categorical features by performing
encoding and these columns have binary value our algorithm will treat them seperately 

- rest of numerical features have different magnitude so we performed scaling on them

```python
final_data.shape
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# identifying the best k value

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

WCSS = []
Silhouette_Score = []

for i in range(2,11):
    kmeans = KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(final_data)
    labels = kmeans.labels_
    WCSS.append(kmeans.inertia_)
    
    score = silhouette_score(final_data,labels)
    Silhouette_Score.append(score)
    
# plotting elbow graph
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(range(2,11),WCSS,color='m',linestyle='--',marker='o')
plt.xlabel('Number of clusters')
plt.ylabel("WCSS")
plt.title("ELbow point graph")


#plotting Silhouette score graph
plt.subplot(1,2,2)
plt.plot(range(2,11),Silhouette_Score,color='m',linestyle='-',marker='o')
plt.xlabel("Cluster numbers")
plt.ylabel("Silhoutte_score")
plt.title("Silhouette score graph")

plt.tight_layout()
plt.show()

    
```

- Optimal value of K is 3

```python
# applying Kmeans on using k value to be 3 based on elbow graph:
kmeans = KMeans(n_clusters=3,init='k-means++')
kmeans.fit(final_data)
```

```python
# predicting cluster number for each data point
label = kmeans.labels_
label
```

```python
print(f"Silhouette score for cluster 3 is: {silhouette_score(final_data,label)}")
```

```python
# Applying Kmeans using k vale to be 4
kmeans = KMeans(n_clusters=4,init='k-means++')
kmeans.fit(final_data)
```

```python
# predicting cluster number for each datapoint
label = kmeans.labels_
label
```

```python
print(f"Silhouette score for cluster 4 is {silhouette_score(final_data,label)}")
```

- It is observed that based on above elbow plot major bend was observed at cluster3 and using 3 cluster we are getting silhouette score is 0.89 
- Using the 4 clusters our silhouette score imporoved that is 0.91 
- fRom this it can be conclude that 4clusters are best representing the datapoints

## Agglomerative approach:(Hierarchial Clustering):

```python
# X is defined
final_data.head()
```

```python
final_data.drop('Cluster',axis=1,inplace=True)
```

```python
X = final_data
```

```python
X.head()
```

```python
# combining/grouping te datapointsand representing it through a dendrogram

import scipy.cluster.hierarchy as sch

plt.figure(figsize=(5,3))
Dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
```

```python
# identifying the k value by visualising the clusters junctions and their distance from dendrogram

plt.figure(figsize=(5,3))
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.title("Dendrogram")
plt.axhline(y=77,color='m',linestyle='--')
plt.axhline(y=36,color='g',linestyle='--')
plt.axhline(y=15,color='r',linestyle='--')


plt.show()
```

- from the analysis we can say that optimal number of cluster is 2

```python
# lets run the algorithm using 2 cluster

from sklearn.cluster import AgglomerativeClustering

agg_cluster = AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')

```

```python
# predicting cluster number for each data point
Y = agg_cluster.fit_predict(X)
Y
```

```python
# converting the array into dataframe
Output = pd.DataFrame(Y,columns=['Agg_cluster_number'])
Output.head(2)
```

```python
Output.tail()
```

```python
# merging the Output df with main dataframe
final_data = pd.concat([X,Output],axis=1)
```
